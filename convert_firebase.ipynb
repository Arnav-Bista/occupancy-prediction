```python
res = []
```

```python
limit = datetime.strptime("2024-05-21", "%Y-%m-%d")
for week_start in week_starts:
    if week_start == "latest" or week_start == "schedule":
        continue
    for points in data["data"][week_start]:
        if points == None:
            continue
        for hour_min in points:
            occupancy = points[hour_min]
            hm = int(hour_min)
            hour = hm // 100
            minute = hm % 100
            timestamp = datetime.strptime(week_start,"%Y-%m-%d")
            timestamp = timestamp.replace(hour = hour, minute = minute)
            if timestamp > limit:
                break
            res.append((timestamp.isoformat(), occupancy))
            
        
    
```
